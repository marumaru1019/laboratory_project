```python
def make_dataset_kai(low_data):
    time_steps = int(input("timesteps:"))
    label_num = int(input("label_num:"))
    data, target = [], []
    maxlen = time_steps

    for i in range(len(low_data)-maxlen):
        data.append(low_data[i:i + maxlen])
        target.append(label_num)

    re_data = np.array(data).reshape(len(data), maxlen,1 )
    re_target = np.array(target).reshape(len(data), 1)
    csv_data = np.array(data).reshape(len(data), maxlen)
    csv_data=pd.DataFrame(csv_data)
    csv_data["LABEL"] =re_target
    

    return re_data, re_target,csv_data
```

```python
#テーブルのデータセット
data_table,label_table,table = make_dataset_kai(y_table)
#椅子のデータセット
data_isu,label_isu,isu = make_dataset_kai(y_isu)
#デニムのデータセット
data_denim,label_denim,denim = make_dataset_kai(y_denim)
#段ボールのデータセット
data_dan,label_dan,dan = make_dataset_kai(y_dan)
#ペットボトルのデータセット
data_petto,label_petto,petto = make_dataset_kai(y_petto)
#指のデータセット
data_yubi,label_yubi,yubi = make_dataset_kai(y_yubi)
#マジックテープのデータセット
data_magic,label_magic,magic = make_dataset_kai(y_magic)
#スポンジのデータセット
data_sponji,label_sponji,sponji = make_dataset_kai(y_sponji)
#ガムテープのデータセット
data_gam,label_gam,gam = make_dataset_kai(y_gam)
#タオルのデータセット
data_taoru,label_taoru,taoru = make_dataset_kai(y_taoru)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input, Model

from tcn import TCN, tcn_full_summary
```

```python
i = Input(shape=(X_train.shape[1], X_train.shape[2]))
optimizer = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

o = TCN(128,nb_stacks=2,dilations=[1, 2, 4, 8,16,32],return_sequences=False)(i)  # The TCN layers are here.
o = Dense(10,activation='softmax')(o)

model = Model(inputs=[i], outputs=[o])
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',  
              metrics = ['accuracy'])

model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', patience=3, verbose=0, mode='auto')
```

```python
history = model.fit(X_train, Y_train, batch_size=1000, epochs=30,validation_split=0.2,callbacks=es)
```

```python
pred = model.predict(X_test)

#予測データ
y_pred = np.argmax(pred,axis=1)

#正解データ
Y_acc = np.argmax(Y_test,axis=1)

class_names = ["table","chair","denim","cardboard","pettobottle","finger","magictape","gamtape","sponge","towel"]

cf_matrix = confusion_matrix(Y_acc, y_pred)
cd_df = pd.DataFrame(cf_matrix)

cd_df.columns=class_names
cd_df.index=class_names

plt.figure(figsize = (24,14))
sns.heatmap(cd_df.astype('float') / cd_df.sum(axis=1), annot=True, fmt='.2%' ,square = True,cmap='Reds')
plt.show()
```
